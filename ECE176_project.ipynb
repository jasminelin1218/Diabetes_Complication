```python
import pandas as pd

cond = pd.read_csv('dataset/conditions.csv')
is_diabetes = cond['DESCRIPTION'].str.contains('Diabetes')
diabetes_records = cond[is_diabetes]
# print(diabetes_records)
diabetes_ids = diabetes_records['PATIENT'].unique()
# print(diabetes_ids)
# print(f"found {len(diabetes_records)} records of diabetes")
# print(f"found {len(diabetes_ids)} of patients with diabetes")

```

```python
df_DiaPatient_list = pd.DataFrame(diabetes_ids, columns=['Patient'])
target_complication = ['Coronary Heart Disease','Diabetic Renal Disease','Diabetic Retinopathy','Neuropathy','Hypertension','Stroke']
for col_name in target_complication:
    matched_ids = cond[cond['DESCRIPTION'].str.contains(col_name, case=False, na=False)]['PATIENT'].unique()
    df_DiaPatient_list[col_name] = df_DiaPatient_list['Patient'].isin(matched_ids).astype(int)
    
print(df_DiaPatient_list)
```
